```python
# List of words
words = ["grandpas", "dads", "are", "rules", "without"]
# A list to hold all possible sentences
sentences = []

# A function that determines all possible permutations of a list of words
def permutations(words):
    sentences = []
    if len(words) <= 1:
        sentences.append(words[0:1])
    else:
        for permutation in permutations(words[1:]):
            for i in range(len(words)):
                sentences.append(permutation[:i] + words[0:1] + permutation[i:])
    return sentences

# Obtain a list of all possible permutations of the list of words
sentences = permutations(words)
```

```python
# Number of Permutations
len(sentences)
```

```python
# Print all permutations
for i in range(len(sentences)):
    string = ""
    for word in sentences[i]:
        string = string + word + " "
    print(str(i+1) + ":   " + string)
```
